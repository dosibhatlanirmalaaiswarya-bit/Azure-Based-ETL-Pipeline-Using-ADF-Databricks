```python
%sql 
create volume retail_data.default.datafiles
```

```python
#Reading data from bronze layer
df_customer = spark.read.parquet("/Volumes/retail_data/default/datafiles/bronze/customers.parquet")
df_stores = spark.read.parquet("/Volumes/retail_data/default/datafiles/bronze/dbo.stores.parquet")
df_products = spark.read.parquet("/Volumes/retail_data/default/datafiles/bronze/dbo.products.parquet")
df_transactions = spark.read.parquet("/Volumes/retail_data/default/datafiles/bronze/dbo.transactions.parquet")
```

```python
display(df_customer)
```

```python

display(df_transactions)
```

```python
from pyspark.sql.functions import col

# Convert types and clean data
df_transactions = df_transactions.select(
    col("transaction_id").cast("int"),
    col("customer_id").cast("int"),
    col("product_id").cast("int"),
    col("store_id").cast("int"),
    col("quantity").cast("int"),
    col("transaction_date").cast("date")
)

df_products = df_products.select(
    col("product_id").cast("int"),
    col("product_name"),
    col("category"),
    col("price").cast("double")
)

df_stores = df_stores.select(
    col("store_id").cast("int"),
    col("store_name"),
    col("location")
)

df_customer = df_customer.select(
    "customer_id", "first_name", "last_name", "email", "city", "registration_date"
).dropDuplicates(["customer_id"])
```

```python
# Join all data
df_silver = df_transactions \
    .join(df_customer, "customer_id") \
    .join(df_products, "product_id") \
    .join(df_stores, "store_id") \
    .withColumn("total_amount", col("quantity") * col("price"))
```

```python
display(df_silver)
```

```python
silver_path = "/Volumes/retail_data/default/datafiles/silver/"

df_silver.write.mode("overwrite").format("delta").save(silver_path)
```

```python
%python
df_silver.createOrReplaceTempView("df_silver")
```

```python
%sql
DROP TABLE IF EXISTS silver_table;
CREATE TABLE default.silver_table
USING DELTA
AS SELECT * FROM df_silver
```

```python
%sql select * from silver_table;
```

```python
# Load cleaned transactions from Silver layer
silver_path = "/Volumes/retail_data/default/datafiles/silver/"
silver_df = spark.read.format("delta").load(silver_path)
```

```python
display(silver_df)
```

```python
from pyspark.sql.functions import sum, countDistinct, avg

gold_df = silver_df.groupBy(
    "transaction_date",
    "product_id", "product_name", "category",
    "store_id", "store_name", "location"
).agg(
    sum("quantity").alias("total_quantity_sold"),
    sum("total_amount").alias("total_sales_amount"),
    countDistinct("transaction_id").alias("number_of_transactions"),
    avg("total_amount").alias("average_transaction_value")
)
```

```python
display(gold_df)
```

```python
gold_path = "/Volumes/retail_data/default/datafiles/gold/"

gold_df.write.mode("overwrite").format("delta").save(gold_path)
```

```python
%python
gold_df.createOrReplaceTempView("gold_df")
```

```python
%sql
CREATE TABLE default.retail_gold_sales_summary
USING DELTA
AS SELECT * FROM gold_df
```
